```python
# RandomForestimport numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import GridSearchCV , RandomizedSearchCV,train_test_split
from sklearn import metrics
from sklearn.model_selection import cross_validate
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

# RandomForest

```python
def r_forest(X,Y):
    model=RandomForestRegressor()
    paramz={'criterion':('mse','mae'),'max_features':('auto','sqrt','log2')}
    
    grid_search = GridSearchCV(estimator = model, param_grid=paramz, scoring=('neg_mean_squared_error')
                               , cv=3, return_train_score=True, iid=False)
    GS_result=cross_validate(grid_search,X,Y,
                                cv=5,scoring=('neg_mean_squared_error'
                                              ,'neg_mean_absolute_error'
                                              ,'neg_median_absolute_error'),return_train_score=True)
    random_search=RandomizedSearchCV(estimator=model,param_distributions=paramz ,scoring=('neg_mean_squared_error')
                                ,cv=3,n_iter=5,return_train_score=True,iid=False)
    RS_result = cross_validate(random_search, x, y, cv=5, scoring=('neg_mean_absolute_error', 'neg_mean_squared_error', 
                                                              'neg_median_absolute_error'), return_train_score=True)
    return GS_result, RS_result
```

# KNeighbors

```python
def kNeighbors(X,Y):
    model=KNeighborsRegressor()
    paramz={'n_neighbors':[5,10,15],'algorithm':('ball_tree','kd_tree','brute')}
    
    grid_search = GridSearchCV(estimator = model, param_grid=paramz, scoring=('neg_mean_squared_error')
                               , cv=3, return_train_score=True, iid=False)
    GS_result=cross_validate(grid_search,X,Y,
                                cv=5,scoring=('neg_mean_squared_error'
                                              ,'neg_mean_absolute_error'
                                              ,'neg_median_absolute_error'),return_train_score=True)
    random_search=RandomizedSearchCV(estimator=model,param_distributions=paramz ,scoring=('neg_mean_squared_error')
                                ,cv=3,n_iter=5,return_train_score=True,iid=False)
    RS_result = cross_validate(random_search, x, y, cv=5, scoring=('neg_mean_absolute_error', 'neg_mean_squared_error', 
                                                              'neg_median_absolute_error'), return_train_score=True)
    return GS_result, RS_result
```

# LinearRegressor

```python
def l_r(X,Y):
    model = LinearRegression()
    paramz = {'fit_intercept' : [True, False],'normalize' : [True, False], 'copy_X' : [True, False] }
    
    grid_search = GridSearchCV(estimator = model, param_grid=paramz, scoring=('neg_mean_squared_error')
                               , cv=3, return_train_score=True, iid=False)
    GS_result=cross_validate(grid_search,X,Y,
                                cv=5,scoring=('neg_mean_squared_error'
                                              ,'neg_mean_absolute_error'
                                              ,'neg_median_absolute_error'),return_train_score=True)
    random_search=RandomizedSearchCV(estimator=model,param_distributions=paramz ,scoring=('neg_mean_squared_error')
                                ,cv=3,n_iter=5,return_train_score=True,iid=False)
    RS_result = cross_validate(random_search, x, y, cv=5, scoring=('neg_mean_absolute_error', 'neg_mean_squared_error', 
                                                              'neg_median_absolute_error'), return_train_score=True)
    return GS_result, RS_result
```

# SuportVectorRegressor

```python
def svr(X,Y):
    model= SVR()
    paramz={'kernel':('linear', 'poly', 'rbf', 'sigmoid')}
    
    grid_search = GridSearchCV(estimator = model, param_grid=paramz, scoring=('neg_mean_squared_error')
                               , cv=3, return_train_score=True, iid=False)
    GS_result=cross_validate(grid_search,X,Y,
                                cv=5,scoring=('neg_mean_squared_error'
                                              ,'neg_mean_absolute_error'
                                              ,'neg_median_absolute_error'),return_train_score=True)
    random_search=RandomizedSearchCV(estimator=model,param_distributions=paramz ,scoring=('neg_mean_squared_error')
                                ,cv=3,n_iter=5,return_train_score=True,iid=False)
    RS_result = cross_validate(random_search, x, y, cv=5, scoring=('neg_mean_absolute_error', 'neg_mean_squared_error', 
                                                              'neg_median_absolute_error'), return_train_score=True)
    return GS_result, RS_result
```

# DecisionTreeRegressor

```python
def dtr(X,Y):
    model=DecisionTreeRegressor()
    paramz={'min_samples_split':[2,3,4]}
    
    grid_search = GridSearchCV(estimator = model, param_grid=paramz, scoring=('neg_mean_squared_error')
                               , cv=3, return_train_score=True, iid=False)
    GS_result=cross_validate(grid_search,X,Y,
                                cv=5,scoring=('neg_mean_squared_error'
                                              ,'neg_mean_absolute_error'
                                              ,'neg_median_absolute_error'),return_train_score=True)
    random_search=RandomizedSearchCV(estimator=model,param_distributions=paramz ,scoring=('neg_mean_squared_error')
                                ,cv=3,n_iter=5,return_train_score=True,iid=False)
    RS_result = cross_validate(random_search, x, y, cv=5, scoring=('neg_mean_absolute_error', 'neg_mean_squared_error', 
                                                              'neg_median_absolute_error'), return_train_score=True)
    return GS_result, RS_result
    
```

```python
def highlight_max(s):
    '''
    highlight the maximum in a Series yellow.
    '''
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in is_max]

def highlight_min(s):
    is_max = s == s.min()
    return ['background-color: blue' if v else '' for v in is_max]
```

# Breast Cancer Dataframe

```python
scaler=MinMaxScaler()
breastC=pd.read_csv("breastC.csv",header=None)
df=breastC.values
y=df[:,0]
scaler.fit(df)
df=scaler.transform(df)
x=df[:,1:]
aux1,aux2=r_forest(x,y)
data=pd.DataFrame()
aux_data1=pd.DataFrame(aux1)
aux_data2=pd.DataFrame(aux2)
data=data.append(aux_data1,ignore_index=True)
data=data.append(aux_data2,ignore_index=True)

aux1,aux2=l_r(x,y)
aux_data1=pd.DataFrame(aux1)
aux_data2=pd.DataFrame(aux2)
data=data.append(aux_data1,ignore_index=True)
data=data.append(aux_data2,ignore_index=True)

aux1,aux2=kNeighbors(x,y)
aux_data1=pd.DataFrame(aux1)
aux_data2=pd.DataFrame(aux2)
data=data.append(aux_data1,ignore_index=True)
data=data.append(aux_data2,ignore_index=True)

aux1,aux2=svr(x,y)
aux_data1=pd.DataFrame(aux1)
aux_data2=pd.DataFrame(aux2)
data=data.append(aux_data1,ignore_index=True)
data=data.append(aux_data2,ignore_index=True)

aux1,aux2=dtr(x,y)
aux_data1=pd.DataFrame(aux1)
aux_data2=pd.DataFrame(aux2)
data=data.append(aux_data1,ignore_index=True)
data=data.append(aux_data2,ignore_index=True)

data=data.abs()
html=data.style.\
    apply(highlight_min).\
    apply(highlight_max).render()
text=open('Cancer.html','w')
text.write(html)
text.close()
data.style.\
    apply(highlight_min).\
    apply(highlight_max)
```

[Suport Vector Regressor](#SuportVectorRegressor)

'''Modelul produs de clasificarea Suport Vector Regressor (așa cum este descris mai sus) depinde doar de un subset al datelor de antrenare, deoarece funcția de cost pentru construirea modelului nu ia in considerare punctele de formare care se află dincolo de margine. În mod analog, modelul produs de Regresia vectorului de suport depinde numai de un subset de date de antrenament, deoarece funcția de cost pentru construirea modelului ignoră orice date de antrenare apropiate de predicția modelului'''

[Linear Regressor](#LinearRegressor)

'''
regresia liniară este o abordare liniară pentru modelarea relației dintre răspunsul scalar (sau variabila dependentă) și una sau mai multe variabile explicative (sau variabile independente). Cazul unei variabile explicative se numește regresie liniară simplă. Pentru mai multe variabile explicative, procesul se numește regresie liniară multiplă. Acest termen este diferit de regresia liniară multivariată, unde sunt prezise variabile multiple dependente  corelate, mai degrabă decât o singură variabilă scalară

LinearRegression se potrivește unui model liniar cu coeficienții w pentru a minimiza suma reziduală de obiecte care apartin anumitei clase dintre răspunsurile observate în setul de date și răspunsurile prevăzute de aproximarea liniară. Matematic rezolvă o problemă a formei# KNieghbors'''
