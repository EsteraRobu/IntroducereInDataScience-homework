```python
import pandas 
import matplotlib.pyplot as plt

nume=['Id','RI','Na','Mg','Al','Si','K','Ca','Ba','Fe','Type']
data =pandas.read_csv('Glass.csv',names=nume,index_col='Id')

data.plot()
```
